1️⃣ Basic Class Creation: Car
------------------------------------------------------------------
Task: Create a class Car with attributes and methods as follows:

Attributes: brand, model, year, mileage

Methods:

drive(distance): Increases mileage by distance.

display_info(): Prints all details of the car.

✅ Enhance: Add a fuel_level attribute and a method refuel(amount) to increase it.
Adjust rest of the attributes accordinly


```python
class Car:
    
    #class varible
    def __init__(self, brand, model, year, mileage, fuel_level=0):
        self.brand = brand
        self.model = model
        self.mileage = mileage
        self.year = year
        self.fuel_level = fuel_level

    #member function
    def drive(self, distance):
        if self.fuel_level >= distance * 0.1: 
            self.mileage += distance
            self.fuel_level -= distance * 0.1 
            print(f"Driven {distance} km. Mileage is now {self.mileage} km.")
        else:
            print("Not enough fuel to drive that distance.")

    def refuel(self, amount):
        self.fuel_level += amount
        print(f"Refueled {amount} liters.\n Current fuel level: {self.fuel_level} liters.")
        
    def display_info(self):
        print("\n")
        print("Car Info:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Mileage: {self.mileage} km")
        print(f"Fuel Level: {self.fuel_level} liters")

       
car1 = Car("Toyota", "Corolla", 2020, 15000)
car1.display_info()
car1.refuel(20)       
car1.drive(100)         
car1.display_info() 
```

```python
class ElectricCar(Car):
    def __init__(self, brand, model, year, mileage, battery_capacity):
        super().__init__(brand, model, year, mileage)
        self.battery_capacity = battery_capacity
        self.charge_level = 100  # Default charge level is 100%

    def charge(self, amount):
        if amount < 0:
            print("Cannot charge with a negative amount.")
            return
        self.charge_level = min(100, self.charge_level + amount)
        print(f"Charged to {self.charge_level}%.")

    def display_info(self):
        super().display_info()
        print(f"Battery Capacity: {self.battery_capacity} kWh, Charge Level: {self.charge_level}%")
```

```python
car = Car("Toyota", "Corolla", 2020, 15000)
car.display_info()
car.drive(300)
car.display_info()

print("\n")

# Creating an ElectricCar object
e_car = ElectricCar("Tesla", "Model 3", 2022, 5000, 75)
e_car.display_info()
e_car.drive(200)
e_car.charge(20)
e_car.display_info()

```
